```python
import pandas as pd
import mysql.connector
import os
import json
```

```python
path_1 = 'D:/Project_2/pulse-master/data/aggregated/transaction/country/india/state/'
```

```python
Agg_state_list=os.listdir(path_1) # in this we have the folders name means every state names 
```

```python
agg_trans = {'States':[], 'Years':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

# here we creating the dict 
```

```python
for i in Agg_state_list:
    p_i = path_1 + i + "/"
    agg_year = os.listdir(p_i)
    
    for year in agg_year:
        a_year = p_i + year + "/"
        agg_year_trans_j = os.listdir(a_year)
        
        for j in agg_year_trans_j:
            y_j = a_year + j
            with open(y_j,'r') as Data:
                file_d = json.load(Data)
           
            
            for z in file_d['data']['transactionData']:
                Name = z['name']
                Count = z['paymentInstruments'][0]['count']
                Amount = z['paymentInstruments'][0]['amount']
                agg_trans['Transaction_type'].append(Name)
                agg_trans['Transaction_count'].append(Count)
                agg_trans['Transaction_amount'].append(Amount)
                agg_trans['States'].append(i)
                agg_trans['Years'].append(year)
                agg_trans['Quarter'].append(int(j.strip('.json')))
    
```

```python
Agg_trans = pd.DataFrame(agg_trans)
```

```python
Agg_trans["States"] = Agg_trans["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Agg_trans["States"] = Agg_trans["States"].str.replace("-"," ")
Agg_trans["States"] = Agg_trans["States"].str.title()
Agg_trans['States'] = Agg_trans['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
```
